## Stackplot ##

```python
x=[1,2,3,4,5]
y1=[0,1,2,3,5]
y2=[0,4,2,6,8]
y3=[1,3,5,7,9]
```

```python
y= np.vstack([y1,y2,y3])
```

```python
labels = ['one', 'two', 'three']
```

```python
fig, ax=plt.subplots()
ax.stackplot(x,y1,y2,y3, labels=labels)
ax.legend(loc='upper left')
```

```python
fig, ax=plt.subplots()
ax.stackplot(x,y);
```
