```python
import numpy as np
```

```python
data = pd.DataFrame([(0.0, np.nan, -1, 1.0),
                    (np.nan, 2.0, np.nan, np.nan),
                    (2.0, 3.0, np.nan, 9.0),
                    (np.nan, 4.0, -4.0, 16.0)],
                   columns = list('abcd'))
```

```python
data
```

```python
data.interpolate(method = 'linear', limit_direction= 'forward', axis=0)
```

```python
data['d'].interpolate(method = 'polynomial', order=2)
```
