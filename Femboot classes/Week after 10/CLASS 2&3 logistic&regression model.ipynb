## Knn

```python
from sklearn.neighbors import KNeighborsClassifier
knn1 = KNeighborsClassifier()
knn1.fit(X,y)
```

## accuracy of the model (classification metric)

```python
knn1.score(X,y) #78% accurate
```

```python
y_pred = knn1.predict(X)
```
